# Q51 モルフォロジー勾配

大津の二値化を行った後、モルフォロジー勾配を求めよ。

モルフォロジー勾配とはモルフォロジー膨張の画像と収縮の画像の差分をとることで、物体の境界線を抽出する手法である。

ここではモルフォロジー処理のN=1とする。

```python
img = cv2.imread("imori.jpg")

gray_img = functions.grayscale(img)
binary_img = functions.ootu_binary(gray_img)

expand_img = functions.morphology_expand(binary_img)
shrinkage_img = functions.morphology_shrinkage(binary_img)
result = expand_img-shrinkage_img

plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```

# Q52 トップハット変換

大津の二値化を行った後、トップハット変換を行え。

トップハット変換とは元画像からオープニング処理を行った画像を差し引いた画像であり、細い線状のものやノイズなどを抽出できると言われる。

ここでは、大津の二値化画像からオープニング処理画像(N=3)を差し引いて求めよ。

```python
n=3
img = cv2.imread("imori.jpg")

gray_img = functions.grayscale(img)
binary_img = functions.ootu_binary(gray_img)

# オープニング処理
open_img = np.copy(binary_img)
for i in range(n):
    open_img = functions.morphology_shrinkage(open_img)
    
for i in range(n):
    open_img = functions.morphology_expand(open_img)
    
result = binary_img-open_img

plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```

# Q53 ブラックハット変換

大津の二値化を行った後、ブラックハット変換を行え。

ブラックハット変換とはクロージング画像から元画像を差し引いた画像であり、これもトップ変換同様に細い線状やノイズを抽出できると言われる。

ここでは、クロージング処理画像(N=3)から大津の二値化画像を差し引いて求めよ。

```python
n=3
img = cv2.imread("imori.jpg")

gray_img = functions.grayscale(img)
binary_img = functions.ootu_binary(gray_img)

# クロージング処理
close_img = np.copy(binary_img)
for i in range(n):
    close_img = functions.morphology_expand(close_img)
    
for i in range(n):
    close_img = functions.morphology_shrinkage(close_img)
    
result = close_img-binary_img

plt.imshow(cv2.cvtColor(result, cv2.COLOR_BGR2RGB))
plt.show()
```

# Q54 テンプレートマッチング SSD

ここではテンプレートマッチングのSSDを用いて、imori_part.jpgがimori.jpgのどこに位置するかをimori.jpgの赤の矩形で図示せよ。

テンプレートマッチングとは、テンプレート画像と全体画像の一部分で類似度が高い位置を探す手法であり、物体検出などで使われる。今では物体検出はCNNで行われるが、テンプレートマッチングは最も基本処理となる。

```python
img = cv2.imread("imori.jpg")
template = cv2.imread("imori_part2.jpg")

H,W,_ = img.shape
Ht,Wt,_ = template.shape

Smin = 255*H*W*3
xsim,ysim= -1,-1
for y in range(H-Ht):
    for x in range(W-Wt):
        S = np.sum((img[y:y+Ht,x:x+Wt]-template)**2)
                
        if S<Smin:
            Smin=S
            xsim,ysim=x,y

fig = plt.figure()
ax = plt.axes()
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
r = patches.Rectangle(xy=(xsim, ysim), width=Ht, height=Wt, ec='#FF0000', fill=False)
ax.add_patch(r)
plt.show()
```

# Q55 テンプレートマッチング SAD

ここではテンプレートマッチングのSADを用いて、imori_part.jpgがimori.jpgのどこに位置するかをimori.jpgの赤の矩形で図示せよ。

SAD(Sum of Absolute Difference)とは画素値の差分の絶対値の和を類似度にする手法であり、Sが最小の位置がマッチング位置となる。

```python
img = cv2.imread("imori.jpg")
template = cv2.imread("imori_part.jpg")

H,W,_ = img.shape
Ht,Wt,_ = template.shape

Smin = 255*H*W*3
xsim,ysim= -1,-1
for y in range(H-Ht):
    for x in range(W-Wt):
        S = np.sum(np.abs((img[y:y+Ht,x:x+Wt]-template)))
                
        if S<Smin:
            Smin=S
            xsim,ysim=x,y

fig = plt.figure()
ax = plt.axes()
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
r = patches.Rectangle(xy=(xsim, ysim), width=Ht, height=Wt, ec='#FF0000', fill=False)
ax.add_patch(r)
plt.show()
```

# Q58 ラベリング 4近傍

seg.pngをラベリングせよ。

ラベリングとは隣接したピクセルに同じラベルを割り当てる作業である。

アルゴリズムは

1. 左上からラスタスキャンを行う。
2. 注目画素i(x,y)が黒画素なら何も行わない。白画素なら、上画素i(x,y-1)と左画素i(x-1,y)に注目し、どちらも0だった場合、最後に割り当てたラベル+1を割り当てる。
3. どちらか一方以上が0でない場合（つまりすでにラベルが割り合っている場合）、上と左に割り当てられたラベルの中で最小の方(0以外)をi(x,y)に割り当てる。ここで、上か左で用いなかったラベルに対応するルックアップテーブルをここで割り当てた番号に変える。
4. 最後、ルックアップテーブルを見て、Sourceに対応する画素に当たる部分をDistinationの値に変換する。

```python
img = cv2.imread("seg.png")

img = img[:,:,0]
result = np.zeros_like(img)
H,W = img.shape
labeling_counter = 0

# ラスタスキャン
for y in range(H):
    for x in range(W):
        if img[y,x]==255:
            up_gaso=0
            left_gaso=0
            # 上下の画素を取り出す
            if y!=0:
                up_gaso = img[y-1,x]

            if x!=0:
                left_gaso = img[y,x-1]

            # どちらも0のとき
            if max(up_gaso,left_gaso)==0:
                labeling_counter+=1
                result[y,x]=labeling_counter
            # どちらかが255のとき
            else:
                left_label=0
                up_label=0
                if x!=0:
                    left_label = result[y,x-1]
                if y!=0:
                    up_label = result[y-1,x]
                
                if left_label==0 and up_label!=0:
                    result[y,x] = up_label
                elif up_label==0 and left_label!=0:
                    result[y,x] = left_label
                elif up_label!=0 and left_label!=0:
                    result[y,x] = min(up_label,left_label)
                    
# 隣接しているノードを統合する
for y in range(H):
    for x in range(W-1):
        left_label = result[y,x]
        right_label = result[y,x+1]
        if min(left_label,right_label)!=0 and left_label!=right_label:
            result[result==right_label] = left_label
            labeling_counter-=1
                    
print(labeling_counter)
label_list = np.unique(result)

COLORS = [[0, 0, 255], [0, 255, 0], [255, 0, 0], [255, 255, 0]]
out = np.zeros((H, W, 3), dtype=np.uint8)

for i in range(labeling_counter):
    out[result == label_list[i+1]] = COLORS[i]
    
plt.imshow(cv2.cvtColor(out, cv2.COLOR_BGR2RGB))
plt.show()
```
