```python
import numpy as np
import matplotlib.pyplot as plt
```

# フーリエ級数展開

$$f(x) = \left\{ \,
    \begin{aligned}
    & 0 (-\pi \leq x < 0) \\
    & x (0 \leq x < \pi) 
    \end{aligned}
\right.$$

```python
def func(x):
    y =x.copy()
    y %=(2*np.pi)
    y -= np.pi
    y[y<0] = 0
    return y

def fourier_func(x,n):
    z =x.copy()
    z %=(2*np.pi)
    z -= np.pi
    y = np.pi/4
    for k in range(1,n+1):
        y+= -(1-((-1)**k))*np.cos(k*z)/(np.pi*(k**2))
        y+= -((-1)**k)*np.sin(k*z)/k
    return y
```

```python
# nを大きくしたとき
x_lines = np.arange(start=-10,stop=10,step=0.01)
y_lines = func(x_lines)

y_fourier_lines_10 = fourier_func(x_lines,10)
y_fourier_lines_20 = fourier_func(x_lines,20)
y_fourier_lines_40 = fourier_func(x_lines,40)

plt.figure(facecolor="white",figsize=(8,8))
plt.subplot(3,1,1)
plt.plot(x_lines,y_lines,label="func")
plt.plot(x_lines,y_fourier_lines_10,label="fourier func N=10")
plt.legend()
plt.subplot(3,1,2)
plt.plot(x_lines,y_lines,label="func")
plt.plot(x_lines,y_fourier_lines_20,label="fourier func N=20")
plt.legend()
plt.subplot(3,1,3)
plt.plot(x_lines,y_lines,label="func")
plt.plot(x_lines,y_fourier_lines_40,label="fourier func N=40")
plt.legend()
plt.show()
```

```python
x_lines = np.arange(start=-10,stop=10,step=0.01)
y_lines = func(x_lines)

plt.figure(facecolor="white",figsize=(8,8))
for k in range(1,4):
    y_fourier_lines = fourier_func(x_lines,k)
    plt.subplot(3,1,k)
    plt.plot(x_lines,y_lines,label="func")
    plt.plot(x_lines,y_fourier_lines,label="fourier func k="+str(k))
    plt.legend()
plt.show()
```
